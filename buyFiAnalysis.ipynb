```python
%matplotlib qt 
import pandas as pd
import glob
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
#import matplotlib.gridspec as gridspec
#import seaborn as sns
#from statsmodels.graphics.mosaicplot import mosaic
```

```python
# Reading the already Processed and merged file
mergedDfFromFile=pd.read_csv('D:\\DataScienceAtNP\\BuyFi\\mergedData.csv', header = 0)
```

```python
mergedDfFromFile['TR_DATE'] = mergedDfFromFile['TR_DATE'].apply(pd.to_datetime , errors='coerce')
uniqueMerchantIds=mergedDfFromFile.merchantId.unique()
resultData =[]
for merchantId in uniqueMerchantIds:
    merchantDf=mergedDfFromFile[mergedDfFromFile['merchantId']==merchantId]
    totTxn=merchantDf['TR_AMOUNT'].sum()
    totTxnInEve=len(merchantDf[merchantDf['TR_DATE'].dt.hour >13])
    totTxnInMorn=len(merchantDf[merchantDf['TR_DATE'].dt.hour <14])
    totTxnValueInEve= (merchantDf[merchantDf['TR_DATE'].dt.hour >13])['TR_AMOUNT'].sum()
    totTxnValueInMorn= (merchantDf[merchantDf['TR_DATE'].dt.hour <14])['TR_AMOUNT'].sum()
    uniqueCustomers=merchantDf.CID.unique()
    meanArr=[]
    for customer in uniqueCustomers :
        custTempDf=merchantDf[merchantDf['CID']==customer]
        mean=custTempDf['TR_AMOUNT'].mean()
        meanArr.append(mean)
    dfMean = pd.DataFrame(meanArr)
    median=dfMean[0].median()
    noOfCustomers=len(uniqueCustomers)
    averageTxn=totTxn/(len(merchantDf))
    tempDf=merchantDf[merchantDf['TR_DATE'].dt.month>0]
    tempDf2=tempDf[merchantDf['TR_DATE'].dt.month<4]
    groupedDf=tempDf2.groupby(tempDf2.TR_DATE.dt.year)
    summ=0
    #doing this so that total quarter value only consists for a year not for all years
    length= len(groupedDf)
    if(length<>0):
        for x in groupedDf.groups.keys():
            y=groupedDf.get_group(x)
            summ=summ+ y['TR_AMOUNT'].sum()
        firstQuarter=summ/length
    else :
        firstQuarter=0
    summ=0
    length=0
    tempDf=merchantDf[merchantDf['TR_DATE'].dt.month>3]
    tempDf2=tempDf[merchantDf['TR_DATE'].dt.month<7]
    groupedDf=tempDf2.groupby(tempDf2.TR_DATE.dt.year)
    length= len(groupedDf)
    if(length<>0):
        for x in groupedDf.groups.keys():
            y=groupedDf.get_group(x)
            summ=summ+ y['TR_AMOUNT'].sum()
        secondQuarter=summ/length
       
    else :
         secondQuarter=0
    summ=0        
    tempDf=merchantDf[merchantDf['TR_DATE'].dt.month>6]
    tempDf2=tempDf[merchantDf['TR_DATE'].dt.month<10]
    groupedDf=tempDf2.groupby(tempDf2.TR_DATE.dt.year)
    length= len(groupedDf)
    if(length<>0):
        for x in groupedDf.groups.keys():
            y=groupedDf.get_group(x)
            summ=summ+ y['TR_AMOUNT'].sum()
        thirdQuarter=summ/length
    else :
         thirdQuarter=0
            
    tempDf=merchantDf[merchantDf['TR_DATE'].dt.month>9]
    tempDf2=tempDf[merchantDf['TR_DATE'].dt.month<13]
    summ=0
    groupedDf=tempDf2.groupby(tempDf2.TR_DATE.dt.year)
    length= len(groupedDf)
    if(length<>0):
        for x in groupedDf.groups.keys():
            y=groupedDf.get_group(x)
            summ=summ+ y['TR_AMOUNT'].sum()
        fourthQuarter=summ/length
    else :
        fourthQuarter=0
    weekdayDf=merchantDf[merchantDf['TR_DATE'].dt.dayofweek<5]
    totTxnOnWeekday=weekdayDf['TR_AMOUNT'].sum()
    weekendDf=merchantDf[merchantDf['TR_DATE'].dt.dayofweek>4]
    totTxnOnWeekend=weekendDf['TR_AMOUNT'].sum()
    noOfWeekendTxn=len(weekendDf)
    noOfWeekdayxn=len(weekdayDf)
    if(noOfWeekendTxn<>0) :
        avgTxnOnWeekend=totTxnOnWeekend/noOfWeekendTxn
    else :
        avgTxnOnWeekend=0
    if(noOfWeekdayxn<>0) :
        avgTxnOnWeekday=totTxnOnWeekday/noOfWeekdayxn
    else :
        avgTxnOnWeekday=0  
    weekdayTxnMedian=weekdayDf['TR_AMOUNT'].median()
    weekendTxnMedian=weekendDf['TR_AMOUNT'].median()
    dataObj={'AvPerCust':median,'MerchantId':merchantId,'TotTxnValue':totTxn,'TotNoOfTxnInMorn':totTxnInMorn,'WeekdayTxnMedian':weekdayTxnMedian,'WeekendTxnMedian':weekendTxnMedian,'TotNoOfTxnInEve':totTxnInEve,'TotTxnValueInMorn':totTxnValueInMorn,'TotTxnOnWeekday':totTxnOnWeekday,'TotTxnOnWeekend':totTxnOnWeekend,'AvgTxnOnWeekday':avgTxnOnWeekday,'AvgTxnOnWeekend':avgTxnOnWeekend,'TotTxnValueInEve':totTxnValueInEve,'NoOfCustomers':noOfCustomers,'AverageTxn':averageTxn,'Q1':firstQuarter,
             'Q2':secondQuarter,'Q3':thirdQuarter,'Q4':fourthQuarter}
    resultData.append(dataObj)
df = pd.DataFrame(resultData)
finalDf=df[['MerchantId','NoOfCustomers','AverageTxn','AvPerCust','TotTxnValue','TotNoOfTxnInEve','TotNoOfTxnInMorn','TotTxnValueInMorn','TotTxnValueInEve','TotTxnOnWeekday','TotTxnOnWeekend','AvgTxnOnWeekday','AvgTxnOnWeekend','WeekendTxnMedian','WeekdayTxnMedian','Q1','Q2','Q3','Q4']]
finalDf.head()
```

```python
#Taking only those merchants whose transactions are more than 2000

mergedDfFromFile1=mergedDfFromFile[mergedDfFromFile['TotTxnValue']>2000]

# adding average values for missing quarters

for index , row in mergedDfFromFile1.iterrows():
    q1=mergedDfFromFile1.ix[index].Q1
    q2=mergedDfFromFile1.ix[index].Q2
    q3=mergedDfFromFile1.ix[index].Q3
    q4=mergedDfFromFile1.ix[index].Q4
    tot=0
    count=0
    if(q1<>0) :
        tot=tot+q1
        count=count+1
    if(q2<>0) :
        tot=tot+q2
        count=count+1
    if(q3<>0) :
        tot=tot+q3
        count=count+1
    if(q4<>0) :
        tot=tot+q4
        count=count+1
    avg=tot/count
    if(q1==0):
        mergedDfFromFile1.set_value(index,'Q1',avg)
    if(q2==0):
        mergedDfFromFile1.set_value(index,'Q2',avg)
    if(q3==0):
        mergedDfFromFile1.set_value(index,'Q3',avg)
    if(q4==0):
        mergedDfFromFile1.set_value(index,'Q4',avg)
mergedDfFromFile1.head()  
        
```
